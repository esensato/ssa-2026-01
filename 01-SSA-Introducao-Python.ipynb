```python
import requests
```

```python
def obter_cep(codigo_cep):
  url = f"https://viacep.com.br/ws/{codigo_cep}/json/"
  print (url)
  return requests.get(url).json()["logradouro"] + ", " + requests.get(url).json()["localidade"]
```

```python
obter_cep("20040-020")
```

```python
df["rua"] = df["cep"].apply(obter_cep)
```

```python
def pesquisa_universidade(nome, pais):
  url = f"http://universities.hipolabs.com/search?name={nome}&country={pais}"
  return requests.get(url).json()
```

```python
for resultado in pesquisa_universidade("Univ", "brazil"):
  print(resultado["name"])
```
